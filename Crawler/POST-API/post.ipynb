```python
import json
import hashlib
import requests  # type: ignore
import os

# Đường dẫn đến thư mục chứa các tệp JSON
directory = r'C:\Users\Tung\Desktop\Chuyển đổi số\tim-viec-main\Crawler\Crawler\Jobs'

# Lấy danh sách tất cả các tệp JSON trong thư mục
txtfiles = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.json')]

for fileName in txtfiles:
    print(f"Processing file: {fileName}")
    
    # Đọc nội dung của tệp JSON với encoding UTF-8
    try:
        with open(fileName, 'r', encoding='utf-8') as f:
            textJson = f.read()
    except UnicodeDecodeError as e:
        print(f"Failed to read {fileName} due to encoding error: {e}")
        continue
        
    # Phân tích cú pháp JSON
    try:
        jobJson = json.loads(textJson)
    except json.JSONDecodeError as e:
        print(f"Failed to parse JSON in {fileName}: {e}")
        continue
    
    # Xây dựng URL API mà không thêm ID công việc
    apiUrl = "https://crawler-job.tung2003ictu.workers.dev/add-job"
    
    # Tạo đối tượng bodyRequest
    bodyRequest = {
        "title": jobJson.get("title"),
        "description": jobJson.get("description"),
        "location": jobJson.get("location"),
        "experience": jobJson.get("experience"),
        "requirement": jobJson.get("requirement"),
        "salary": jobJson.get("salary"),
        "expiration_date": jobJson.get("expiration_date"),
        "job_type": jobJson.get("job_type"),
        "created_at": jobJson.get("created_at"),
        "url": jobJson.get("url"),
        "company": jobJson.get("company")
    }
    
    headers = {
        'Content-Type': 'application/json',
    }

    print(f"Posting to: {apiUrl}")
    print("Request body:", json.dumps(bodyRequest, indent=2))
    
    # Gửi yêu cầu POST đến API
    try:
        response = requests.post(apiUrl, json=bodyRequest, headers=headers)
        print("Response status code:", response.status_code)
        print("Response:", response.text)
    except requests.exceptions.RequestException as e:
        print("Request failed:", e)
```
